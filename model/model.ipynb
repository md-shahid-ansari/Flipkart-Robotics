```python
import cv2
from matplotlib import pyplot as plt
import os
```

```python

# Path to the video file
video_path = '/content/BB_2d6a1f57-5224-4b1f-8207-88c4899537a8_preview.mp4'  # Replace with your video file path

# Check if the video file exists
if not os.path.exists(video_path):
    print("Video file does not exist.")
    exit()

# Create a directory to save frames if it doesn't exist
output_dir = '/content/images'
os.makedirs(output_dir, exist_ok=True)

# Open the video file
video = cv2.VideoCapture(video_path)

# Check if video opened successfully
if not video.isOpened():
    print("Error: Could not open video.")
    exit()

frame_count = 0
saved_frames = 0

# Loop through the video frames
while True:
    ret, frame = video.read()

    if not ret:
        break  # End of video

    # Save every 30th frame (adjust if you want more/less frequent frames)
    if frame_count % 30 == 0:
        frame_filename = f"{output_dir}/frame_{saved_frames}.jpg"
        cv2.imwrite(frame_filename, frame)
        print(f"Saved frame: {frame_filename}")  # Print saved frame info
        saved_frames += 1

    frame_count += 1

# Release the video capture object
video.release()

print(f"Extracted {saved_frames} frames and saved them in '{output_dir}' directory.")

# List the saved frames
print("Files in output directory:")
print(os.listdir(output_dir))
```

```python
# Path to the images directory
images_dir = '/content/images'

# Get a list of image file paths in the specified directory
image_files = [os.path.join(images_dir, f) for f in os.listdir(images_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Loop through and display images
for image_path in image_files:
    image = cv2.imread(image_path)

    if image is not None:  # Check if the image is loaded successfully
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Display the image
        plt.imshow(image_rgb)
        plt.axis('off')
        plt.title(f"Image: {os.path.basename(image_path)}")
        plt.show()
    else:
        print(f"Error loading image: {image_path}")
```
